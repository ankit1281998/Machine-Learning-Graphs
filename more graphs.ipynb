```python
Y=[2,3,4,5,6]
Y2=[3,6,3,6,7]
Y3=[2,6,7,2,3]
```

```python
flights=sns.load_dataset("flights").pivot("month","year","passengers")
flights
```

```python
from urllib.request import urlretrieve
urlretrieve("https://i.imgur.com/SkPbq.jpg","image.jpg")
```

```python
from PIL import Image
img=Image.open("image.jpg")
```

```python
plt.imshow(img)
```

```python
plt.subplots(2,3)
```

```python
fig,axes=plt.subplots(2,3,figsize=(12,9))

plt.tight_layout(pad=2)
```

```python
fig,axes=plt.subplots(2,3,figsize=(12,9))
axes[0,0].plot(Y,Y2,"o--r")
axes[0,0].plot(Y,Y3,"s--b")
axes[0,0].set_title("graph1")
axes[0,0].set_xlabel("x")
axes[0,0].set_xlabel("y")
axes[0,0].legend(["x","y"])

axes[0,1].set_title("graph2")
sns.scatterplot(x=Y,y=Y2,hue=Y3,s=100,ax=axes[0,1])

axes[0,2].set_title("graph3")
axes[0,2].hist([Y2,Y3],stacked=True)
axes[0,2].legend(["Y2","Y3"])

axes[1,0].set_title("graph4")
axes[1,0].bar(Y,Y3)

axes[1,1].set_title("graph5")
sns.heatmap(flights,cmap="Blues",ax=axes[1,1])

axes[1,2].set_title("graph6")
axes[1,2].imshow(img)
axes[1,2].axis("off")
#axes[1,2].set_xticks([])
#axes[1,2].set_yticks([])

plt.tight_layout(pad=2)

```
