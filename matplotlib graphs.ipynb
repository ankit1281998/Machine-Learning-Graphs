```python
#Bar Plot
Y=np.random.rand(5)
Y2=np.random.rand(5)
```

```python
plt.figure(figsize=(12,4))
barWidth=0.3

plt.bar(np.arange(len(Y)),Y,width=barWidth,color="blue")
```

```python
plt.figure(figsize=(12,4))
barWidth=0.3

plt.bar(np.arange(len(Y)),Y,width=barWidth,color="blue",label="Y")

plt.bar(np.arange(len(Y2)),Y2,width=barWidth,color="orange",label="Y2")
plt.legend()
#plt.show()
```

```python
plt.figure(figsize=(12,4))
barWidth=0.3

plt.bar(np.arange(len(Y)),Y,width=barWidth,color="blue",label="Y")

plt.bar(np.arange(len(Y2)),Y2,width=barWidth,color="orange",bottom=Y,label="Y2")
plt.legend()
plt.show()
```
